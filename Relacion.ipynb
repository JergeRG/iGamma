```python
import re
import csv
```

```python
listF = []
with open('entitiesFiltradas.csv') as file:
    reader = csv.reader(file)
    for row in reader:
        listF.append(row)
```

```python
for i in range(1,len(listF)):
    listF[i][1] = "entitie"+str(i)
```

```python
listF
```

```python
with open('listRelacion.csv', 'w') as file:
    writer = csv.writer(file)
    writer.writerows(listF)
```
